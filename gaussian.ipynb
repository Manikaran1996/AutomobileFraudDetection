```python
from math import sqrt, exp, pi
def prob(x, m, s):
   
    """p = numpy.zeros(x.shape)
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            p[i,j] = 1/(sqrt(2*pi)*s * numpy.exp(((x[i][j]-m)**2)/(2*s*s)))"""
    
    p = (1/sqrt(2*pi)*s) * numpy.exp(-((x-m)*(x-m))/(2*s*s))
    
    return p
```
